```python
import numpy as np
from matplotlib import pyplot as plt

data_mean = []
for i in range(0,2200,200):
    alldata = np.loadtxt("/Users/sholafolarin/Desktop/DOT_FILES/increase_Tape_10k/iTape_{}_10k/piezo_loc.dat".format(i))
    data_mean.append(alldata)
    

num = np.zeros((11,6))

for i in range(0,11):
    for j in range(4,10):
        num[i,j-4] = np.mean(data_mean[i][:,j])
        #mydat.append([i,j,mydata])
        
E = [0,200,400,600,800,1000,1200,1400,1600,1800,2000]
nrow = 2
ncol =3

fig,ax = plt.subplots(nrow, ncol,  figsize=(15,7), sharex=True)

mycolor = ['r','b','g','orange', 'magenta', 'r']
labels = ['Strain1', 'Strain2', 'Strain3', 'Strain4', 'Strain5', 'Strain6']
for i, a in enumerate(ax.flatten()):
    a.plot(E, num[:,i], '-o', color =mycolor[i], label= labels[i])
    fig.suptitle("Strain as a function of Electric Field", fontsize =20)
    a.legend()
    fig.supxlabel('Electric Field kV/cm', fontsize =20)
    fig.supylabel('Strain', fontsize =20)
    
# fig.text(0.5, 0.04, 'Electric Field (kV/cm)', ha='center', fontsize =20)
# fig.text(0.04, 0.5,  'Strain',va='center', rotation='vertical', fontsize =20)
    

            
       
```
