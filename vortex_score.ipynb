```python
import pandas as pd
from matplotlib import pyplot as plt

sheet_name = ['XY', 'XZ','YZ', 'dXY', 'dXZ','dYZ']

lst = []

for i in sheet_name:

    df = pd.read_excel("/Users/sholafolarin/Desktop/DOT_FILES/vortex_score.xlsx", sheet_name=i)
    lst.append(df)
    
    
    
```

```python
from matplotlib import pyplot as plt
fields = ['0kv/cm', '600kv/cm', '1400kv/cm']

ncol = 3
nrow = 2
fig, ax = plt.subplots(nrow, ncol, figsize = (15,8))

mycolor = ['r','b','g','orange', 'magenta', 'r']
labels = ['XY', 'XZ', 'YZ', 'XY', 'XZ', 'YZ']
            
for i, a in enumerate(ax.flatten()):
    for j in fields:
        a.plot( lst[i-6]['plane'], lst[i-6][j], '-o', label=j )
        a.set_title(labels[i])
        fig.supxlabel('Number of Planes', fontsize =20)
        fig.supylabel('Number of Vortices', fontsize =20)
a.legend(loc='center left', bbox_to_anchor=(1, 1))
```
